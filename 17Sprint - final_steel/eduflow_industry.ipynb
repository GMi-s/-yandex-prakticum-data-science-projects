## Изучение характеристик процесса обработки стали

### Определение длительности времени между первым и последним замером температуры. 

*Таблица `data_temp` содержит данные о последовательных измерениях температуры*

```python
data_temp.head(5)
```

```python
#Удалим пропуски
data_temp = data_temp.dropna()

#Переведём время в формат DateTime
data_temp['Время замера'] = pd.to_datetime(data_temp['Время замера'])
```

```python
#Сгруппируем замеры таким образом, чтобы сохранить только последнее измерение температуры
data_temp_last = data_temp.groupby(by=['key'], as_index = False).nth(-1)
data_temp_last.rename(columns={'Температура': 'Последняя температура'}, inplace=True)
data_temp_last.rename(columns={'Время замера': 'Время последней температуры'}, inplace=True)
data_temp_last.head(5)
```

```python
#Сгруппируем замеры таким образом, чтобы сохранить только первое измерение температуры
data_temp_first = data_temp.groupby(by=['key'], as_index = False).nth(0)
data_temp_first.rename(columns={'Температура': 'Первая температура'}, inplace=True)
data_temp_first.rename(columns={'Время замера': 'Время первой температуры'}, inplace=True)
data_temp_first.head(5)
```

```python
#Объединим данные в одну таблицу
data_temp_group = data_temp_first.join(data_temp_last.set_index('key'), on = 'key', how = 'inner')
data_temp_group.head(5)
```

```python
data_temp_group['Длительность между первым и последним замером температуры'] = (data_temp_group['Время последней температуры'] \
                                        - data_temp_group['Время первой температуры']).dt.total_seconds()

data_temp_group.head(5)
```

```python
data_temp_group = data_temp_group.drop(['Время первой температуры','Первая температура',
                                        'Время последней температуры','Последняя температура'], axis=1)
data_temp_group.head(5)
```

#### Вычисление статистики длительности времени между первым и последним замером температуры: средние, минимальные и максимальные значения, медианы и величины 25%- и 75%-квартилей

```python
def statistic_info (df,column):
    '''
    Принимает на вход источник данных (название таблицы) и категорию (название столбца)
    Возвращает: среднее значение, минимум, максимум, медиану, значние 25%- и 75%-квантилей
    '''
    
    mean = round(df[column].mean(),2)
    minimum = round(df[column].min(),2)
    maximum = round(df[column].max(),2)
    median = round(df[column].median(),2)
    q_25 = round(df[column].quantile(0.25),2)
    q_75 = round(df[column].quantile(0.75),2)
    
    return mean, minimum, maximum, median, q_25, q_75
```

```python
#Вычисление среднего значения
#Вычисление минимального значения
#Вычисление максимального значения
#Вычисление медианы
#Вычисление значения 25%-квантиля
#Вычисление значения 75%-квантиля
time_temp_mean, time_temp_min, time_temp_max, time_temp_median, time_temp_q_25, time_temp_q_75  = \
        statistic_info (data_temp_group,'Длительность между первым и последним замером температуры')

print('Среднее значение:',time_temp_mean)
print('Минимальное значение:',time_temp_min)
print('Максимальное значение:',time_temp_max)
print('Медиана:',time_temp_median)
print('Значение 25%-квантиля:',time_temp_q_25)
print('Значение 25%-квантиля:',time_temp_q_75)
```

### Определение суммарного времени нагрева электродами, то есть суммы значений по всем промежуткам между запусками нагрева электродов

*Таблица `data_arc` содержит данные о последовательных нагревах электродами*

```python
data_arc.head(5)
```

```python
#Переведём время в формат DateTime
data_arc['Начало нагрева дугой'] = pd.to_datetime(data_arc['Начало нагрева дугой'])
data_arc['Конец нагрева дугой'] = pd.to_datetime(data_arc['Конец нагрева дугой'])
```

```python
#Создадим столбец, где будет храниться время нагрева дугой
data_arc['Время нагрева дугой'] = (data_arc['Конец нагрева дугой'] - \
                                   data_arc['Начало нагрева дугой']).dt.total_seconds()
data_arc.head(5)
```

```python
#Сгруппируем данные по ключу с суммированием времени нагрева
data_arc_heat_time = data_arc[['key','Время нагрева дугой']].groupby(by=['key'], as_index = False).sum()
data_arc_heat_time.head(5)
```

#### Вычисление статистики суммарного времени нагрева электродами: средние, минимальные и максимальные значения, медианы и величины 25%- и 75%-квартилей

```python
#Вычисление среднего значения
#Вычисление минимального значения
#Вычисление максимального значения
#Вычисление медианы
#Вычисление значения 25%-квантиля
#Вычисление значения 75%-квантиля
time_heat_mean, time_heat_min, time_heat_max, time_heat_median, time_heat_q_25, time_heat_q_75  = \
        statistic_info (data_arc_heat_time,'Время нагрева дугой')

print('Среднее значение:',time_heat_mean)
print('Минимальное значение:',time_heat_min)
print('Максимальное значение:',time_heat_max)
print('Медиана:',time_heat_median)
print('Значение 25%-квантиля:',time_heat_q_25)
print('Значение 25%-квантиля:',time_heat_q_75)
```

### Определение количества запусков нагрева электродами

```python
#Вычислим и добавим в таблицу количество нагревов
data_arc_heat_counts = data_arc.groupby(['key']).size().reset_index(name='Количество нагревов')
data_arc_heat_counts.head(5)
```

#### Вычисление статистики количества запусков нагрева электродами: средние, минимальные и максимальные значения, медианы и величины 25%- и 75%-квартилей

```python
#Вычисление среднего значения
#Вычисление минимального значения
#Вычисление максимального значения
#Вычисление медианы
#Вычисление значения 25%-квантиля
#Вычисление значения 75%-квантиля
heat_counts_mean, heat_counts_min, heat_counts_max, heat_counts_median, heat_counts_q_25, heat_counts_q_75  = \
        statistic_info (data_arc_heat_counts,'Количество нагревов')

print('Среднее значение:',heat_counts_mean)
print('Минимальное значение:',heat_counts_min)
print('Максимальное значение:',heat_counts_max)
print('Медиана:',heat_counts_median)
print('Значение 25%-квантиля:',heat_counts_q_25)
print('Значение 25%-квантиля:',heat_counts_q_75)
```

### Определение среднего соотношение потребления активной и реактивной мощности

```python
#Сгруппируем данные по ключу с вычислением среднего по соотношению мощностей
data_arc_power_rate = data_arc[['key','Соотношение мощностей']].groupby(by=['key'], as_index = False).mean()
data_arc_power_rate.head(5)
```

#### Вычисление статистики среднего соотношение потребления активной и реактивной мощности: средние, минимальные и максимальные значения, медианы и величины 25%- и 75%-квартилей

```python
#Вычисление среднего значения
#Вычисление минимального значения
#Вычисление максимального значения
#Вычисление медианы
#Вычисление значения 25%-квантиля
#Вычисление значения 75%-квантиля
power_rate_mean, power_rate_min, power_rate_max, power_rate_median, power_rate_q_25, power_rate_q_75  = \
        statistic_info (data_arc_power_rate,'Соотношение мощностей')

print('Среднее значение:',power_rate_mean)
print('Минимальное значение:',power_rate_min)
print('Максимальное значение:',power_rate_max)
print('Медиана:',power_rate_median)
print('Значение 25%-квантиля:',power_rate_q_25)
print('Значение 25%-квантиля:',power_rate_q_75)
```
