## Разбивка данных по выборкам

### Признаки 'features' и целевой признак 'target'

```python
features = users_behavior_data.drop(['is_ultra'], axis=1)  #извлекаем признаки
target = users_behavior_data['is_ultra']                   #извлекаем целевой признаки
```

### Выделение тестовой выборки из общей

*После создания 'features' и 'target' необходимо разделить выборку на треннировочную, валидационную и тестовую. Разделим их в соотношении 3/5 - 1/5 - 1/5, соответственно.* <br>
*Для этого последовательно применим 'train_test_split' сначала для разделения на треннировочную и тестовую выборки, а затем выделим из тестовой выборки валидационную.*

```python
#Разделим общую выборку на целевую и тестовую в соотношении 60% на 40%, т.к. 
#далее разделим тестовую выборку на тестовую и валидационную
features_train, features_test, target_train, target_test = train_test_split(
    features, target, train_size=0.60,random_state=12345)

print(f'Размер массива целевых значений обучающей выборки:{target_train.shape}')
print(f'Размер массива целевых значений тестовой выборки: {target_test.shape}')
print(f'Размер обучающей выборки: {features_train.shape}')
print(f'Размер тестовой выборки: {features_test.shape}')
```

### Выделение валидационной выборки из тестовой

```python
#Разделим тестовую выборку на тестовую и валидационную
features_test, features_valid, target_test, target_valid = train_test_split(
    features_test, target_test, train_size=0.50,random_state=12345)

print(f'Размер массива целевых значений тестовой выборки: {target_test.shape}')
print(f'Размер массива целевых значений валидационной выборки:{target_valid.shape}')
print(f'Размер тестовой выборки:{features_test.shape}')
print(f'Размер валидационной выборки:{features_valid.shape}')
```

**Вывод:** <br>
- Из общих данных были выделены признаки 'futures' и 'target' <br>
- И общей совокупности данных были выделены три выборки:
  - обучающая ('features_train' и 'target_train') размером 1928 строк <br>
  - валидационная ('features_valid' и 'target_valid') размером 643 строки <br>
  - тестовая ('features_test' и 'target_test') размером 643 строки

### Дерево решений

В качестве гиперпараметра возьмём глубину дерева.<br>
Максимальную глубину дерева будем исследовать в диапазоне от 2 до 10.

```python
%%time
best_model_tree = None
best_depth_tree = 0
best_result_tree = 0
for depth in range (2,10):
    model = DecisionTreeClassifier(random_state=12345,max_depth=depth)
    model.fit(features_train,target_train)
    predictions_valid = model.predict(features_valid)
    result = accuracy_score(target_valid, predictions_valid)
    if result > best_result_tree:
        best_model_tree = model
        best_result_tree = result
        best_depth_tree = depth
        joblib.dump(best_model_tree, 'best_model_tree.joblib')

print("Глубина дерева:", best_depth_tree)
print("Лучшее значение accuracy:",best_result_tree.round(4))
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Подбор парамтеров был сделан верно.

### Случайный лес

В качестве гиперпараметра возьмём n_estimators(количество деревьев в лесу)

```python
%%time

best_model_forest = None
best_result_forest = 0
for est in range(1, 51):
    model = RandomForestClassifier(random_state=12345, n_estimators=est)
    model.fit(features_train,target_train) 
    result = model.score(features_valid,target_valid) 
    if result > best_result_forest:
        best_model_forest = model
        best_result_forest = result
        joblib.dump(best_model_forest, 'best_model_forest.joblib')

print("Количество деревьев наилучшей модели Случайный лес:",est)
print("Accuracy наилучшей модели на валидационной выборке:", best_result_forest.round(4))
```

<div class="alert alert-block alert-info">
<b>Совет: </b> Тут можно было настроить сразу несколько параметров.
</div>

<div class="alert alert-block alert-warning">
<b>Изменения:</b> Были внесены следующие изменения - Скопировал блок и перенёс его ниже, добавив параметры min_samples_split (Минимальное количество примеров для разделения) и min_samples_leaf (Минимальное количество объектов в листе) из соответстующего раздела тренажёра

```python
%%time

best_model_forest = None
best_result_forest = 0
for i,split in enumerate(range(2,10)):
    for j,leaf in enumerate(range(1,4)):
        model_forest = RandomForestClassifier(random_state=12345, min_samples_split=split,min_samples_leaf=leaf) 
        model_forest.fit(features_train,target_train)
        
        result_forest = model_forest.score(features_valid,target_valid) 
        if result_forest > best_result_forest:
            best_model_forest = model_forest
            best_result_forest = result_forest
print("Accuracy наилучшей модели на валидационной выборке:",best_result_forest)  
print("Минимальное количество примеров для разделения:",split)
print("Минимальное количество объектов в листе:",leaf)
        
best_model_forest = None
best_result_forest = 0        
for est in range(1, 111):
    model = RandomForestClassifier(random_state=12345,min_samples_split=9,min_samples_leaf=3,n_estimators=est)
    model.fit(features_train,target_train)
    result = model.score(features_valid,target_valid)
    if result > best_result_forest:
        best_model_forest = model
        best_result_forest = result
        joblib.dump(best_model_forest, 'best_model_forest.joblib')

print("Количество деревьев наилучшей модели Случайный лес:",est)
print("Accuracy наилучшей модели на валидационной выборке:", best_result_forest)
```
