```python
def corr_def (df,threshold):
    '''
    Возвращает признаки с очень высокой корреляцией >= 0,9 и на выходе создает два списка: 
    col_corr_all - собраны все уникальные столбцы с высокой корреляцией;
    col_corr_del - каждый первый из взаимокоррелирующих
    '''
    col_corr_del = []
    col_corr_all = []
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if  (corr_matrix.iloc[i, j] >= threshold):
                if (corr_matrix.columns[j],corr_matrix.columns[i] not in col_corr_all):
                    col_name_i = corr_matrix.columns[i]
                    col_name_j = corr_matrix.columns[j]
                    if col_name_i not in col_corr_all:
                        col_corr_all.append(col_name_i)
                        if col_name_j not in col_corr_all:
                            col_corr_all.append(col_name_j)
                if corr_matrix.columns[j] not in col_corr_del:
                    col_name = corr_matrix.columns[i]
                    col_corr_del.append(col_name)

    return col_corr_all, col_corr_del
```

### Расчёт эффективности добычи золота

*По условию задания нужно проверить, что эффективность обогащения рассчитана правильно. Для этого воспользуемся формулой:*<br>

$$Recovery = \frac{C*\left ( F - T\right)}{F*\left ( C - T\right)}*100\%$$

*где:* <br>

•   ***Recovery*** - *эффективность обогащения;*  
•	***C*** — *доля золота в концентрате после флотации/очистки;*  
•	***F*** — *доля золота в сырье/концентрате до флотации/очистки;*  
•	***T*** — *доля золота в отвальных хвостах после флотации/очистки.*  

*А так же вычислим её на обучающей выборке для признака `rougher.output.recovery`. Найдем MAE (среднюю абсолютную ошибку) между нашими расчётами и значением признака.*

```python
gr_train_calc = gr_train.copy(deep=True)
```

```python
def calc_recovery(df,color):
    '''
    Функция расчета средней абсолютной ошибки, 
    для проверки правильности расчетов эффективности обогащения.
    '''
    df.dropna(inplace=True)
    C = df['rougher.output.concentrate_au']
    F = df['rougher.input.feed_au'] 
    T = df['rougher.output.tail_au']
    df['rougher.output.recovery_calc'] = ((C*(F-T))/(F*(C-T)))*100
    mae = mean_absolute_error(df['rougher.output.recovery'],df['rougher.output.recovery_calc'])
    df.drop('rougher.output.recovery_calc', axis=1,inplace=True)
    display(HTML(f"<font color={color}> Cредняя абсолютная ошибка (MAE): {mae} </font>"))
    return mae
```

### Сравнение распределения гранул сырья на обучающей и тестовой выборках

```python
def compare_granules(train,test):
    
    '''
    Функция выводит гистограммы, подаваемых в аргумент данных, 
    а так же сравнивает их средний размер, дисперсию и мажквартильный размах.
    '''
    
    fig = plt.figure(figsize=(15, 6))

    ax1 = plt.subplot(1, 2, 1)
    ax1 = sns.histplot(train, kde=True,bins=60,color = 'green') 

    ax1.set(xlabel ='Размеры гранул', ylabel = 'Количество гранул', 
            title ='Распределение размеров гранул сырья в обучающей выборке')


    ax2 = plt.subplot(1, 2, 2)
    ax1 = sns.histplot(train,  kde=True,bins=60,color = 'purple') 

    ax2.set(xlabel ='Размеры гранул', ylabel = 'Количество гранул',
            title ='Распределение размеров гранул сырья в тестовой выборке')
    plt.show()
    
    # расчет средних
    mean_train = train.mean()
    mean_test = test.mean()
    display(HTML(f"<font color='green'>Средний размер гранул в обучающей выборке: {mean_train}</font>"))
    display(HTML(f"<font color='purple'>Средний размер гранул в тестовой выборке: {mean_test}</font>"))
    print(' ')
    
    # расчет дисперсии
    var_train = np.var(train, ddof=1)
    var_test = np.var(test, ddof=1)
    display(HTML(f"<font color='green'>Дисперсия размеров гранул в обучающей выборке: {var_train}</font>"))
    display(HTML(f"<font color='purple'>Дисперсия размеров гранул в тестовой выборке: {var_test}</font>"))
    print(' ')
    
    # расчет стандартного отклонения
    std_train = np.sqrt(var_train)
    std_test = np.sqrt(var_test)
    display(HTML(
        f"<font color='green'>Стандартное отклонение размеров гранул в обучающей выборке: \
        {std_train}</font>"))
    display(HTML(
        f"<font color='purple'>Стандартное отклонение размеров гранул в тестовой выборке: \
        {std_test}</font>"))
    print(' ')
    
    # расчет межквартильного размаха 
    adv_train_min = mean_train - 3*std_train
    adv_train_max = 3*std_train + mean_train
    adv_test_min = mean_test - 3*std_test
    adv_test_max = 3*std_test + mean_test
    display(HTML(
        f"<font color='green'>Размеры большей части гранул в обучающей выборке колеблются в промежутке от:\
        {adv_train_min} до {adv_train_max}</font>"))
    display(HTML(
        f"<font color='purple'>Размеры большей части гранул в обучающей выборке колеблются в промежутке от:\
        {adv_test_min} до {adv_test_max}</font>"))
```

### Метрика качества

*Метрика качества, необходимая для оценки эффективности процесса очистки золотосодержащей руды — sMAPE (симметричное среднее абсолютное процентное отклонение).  
Метрика sMAPE вычисляется так:*


$$sMAPE = \frac{1}{N} \sum_{i=1}^{N}\frac{\left | y_{i} -\
\hat{y}_{i}\right |}{\left ( \left | y_{i} \right | +\left | \hat{y}_{i} \right |\right )/2} *100\%$$  

*Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество:* $$y_{i}$$   
*Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке:* $$\hat{y}_{i}$$  
*Количество объектов в выборке:* $$N$$  
*Суммирование по всем объектам выборки (i меняется от 1 до N).* $$\sum_{i=1}^{N}$$

```python
def smape_calc(target,prediction):
    '''
    Функция вычисления метрики smape
    '''
    diff = abs(target - prediction)
    add = (abs(target) + abs(prediction))/2
    smape_stage = 1/len(target)*sum(diff/add)*100
    
    return smape_stage
```

```python
def final_smape(smape_r,smape_f):
    '''
    Функция вычисления итоговой метрики smape
    '''
    final_smape = 0.25*smape_r + 0.75*smape_f
    return final_smape
```

```python
def random_grid(model,params,scoring,features,target):
    '''
    Функция осуществляет грубый подбор параметров, сужая их поиск, из диапазона переданного в аргумент, 
    для принимаемой в аргумент модели
    '''
    
    model_random = RandomizedSearchCV(model,params,cv=5, n_jobs=-1,scoring=score)
    model_random.fit(features,target)
    best_params_random = model_random.best_params_
    best_score_random = model_random.best_score_
    return model_random,best_params_random,best_score_random
```

```python
def params_grid(model,params,scoring,features,target):
    '''
    Функция ищет лучшие параметры из диапазона переданного в аргумент, 
    для принимаемой в аргумент модели
    '''
    model_grid = GridSearchCV(model,params,cv=5, n_jobs=-1,scoring=score)
    model_grid.fit(features,target)
    best_params = model_grid.best_params_
    best_score = model_grid.best_score_
    return model_grid,best_params,best_score
```
