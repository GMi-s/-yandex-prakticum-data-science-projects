```python
def missed_data(df):
    '''
    –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∏–º—è DataFrame.
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
    '''
    for col in df.columns:
        data_missed = np.mean(df[col].isna())
        print('{} - {}%'.format(col, round(data_missed*100,2)))
```

### –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤  –≤ —á–∏—Å–ª–µ–Ω–Ω—ã–µ (One-Hot Encoding, OHE)

### –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞

```python
features = churn_data_ohe.drop('exited', axis=1)
target = churn_data_ohe['exited']
```

### –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é, —Ç–µ—Å—Ç–æ–≤—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏

```python
# –†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ 40%/60%  –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
features_train, features_test, target_train, target_test = train_test_split(
    features, target, train_size=0.60,random_state=12345)

# –†–∞–∑–¥–µ–ª–∏–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É –Ω–∞ —Ç–µ—Å—Ç–æ–≤—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –ø–æ—Ä–æ–≤–Ω—É 20%/20% –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏
features_test, features_valid, target_test, target_valid = train_test_split(
    features_test, target_test, train_size=0.50,random_state=12345)

display(HTML(f"<font color='red'>–¶–µ–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –æ–±—É—á–∞—é—â–µ–π, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π —Å–±–æ—Ä–æ–∫: \
{target_train.shape} {target_valid.shape} {target_test.shape}</font>"))
display(HTML(f"<font color='blue'>–ü—Ä–∏–∑–Ω–∞–∫–∏ –æ–±—É—á–∞—é—â–µ–π, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π —Å–±–æ—Ä–æ–∫: \
{features_train.shape} {features_valid.shape} {features_test.shape}</font>"))
```

<div class="alert alert-success">
<b>üëç –£—Å–ø–µ—Ö:</b> –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–±–∏—Ç—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ
</div>

### –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
# –°–æ–∑–¥–∞–¥–∏–º –æ–±—ä–µ–∫—Ç –¥–ª—è —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ StandardScaler
scaler = StandardScaler()           

# –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –ø–æ–¥–ª–µ–∂–∞—â–∏—Ö –º–∞—Å—à–∞–±–∏—Ä–æ–≤–∞–Ω–∏—é (–≤—Å–µ —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –±–∏–Ω–∞—Ä–Ω—ã—Ö)
numeric = ['creditscore', 'age', 'tenure', 'balance','numofproducts','hascrcard','estimatedsalary']

# –ü—Ä–æ–∏–∑–≤–µ–¥—ë–º –Ω–∞—Å—Ç—Ä–æ–π–∫—É –Ω–∞ –¥–∞–Ω–Ω—ã—Ö –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
scaler.fit(features_train[numeric]) 

features_train[numeric] = scaler.transform(features_train[numeric])     # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É
features_valid[numeric] = scaler.transform(features_valid[numeric])     # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É
features_test[numeric] = scaler.transform(features_test[numeric])       # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É

features_train.head()
```

<div class="alert alert-success">
<b>üëç –£—Å–ø–µ—Ö:</b> –¢—ã –Ω–µ –∑–∞–±—ã–ª –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –ø—Ä–∏–º–µ–Ω–∏–ª –µ–≥–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ. –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –æ–±—É—á–µ–Ω–æ –Ω–∞ —Ç—Ä–µ–π–Ω–µ –∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–æ –Ω–∞ –≤—Å–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ. –¢–æ–ª—å–∫–æ –Ω–∞ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö.
</div>

### –í—ã–≤–æ–¥—ã

*–î–∞–Ω–Ω—ã–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω—ã –∫ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—é*  
* *–ò–∑–≤–ª–µ—á–µ–Ω—ã —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –∏ –ø—Ä–∏–∑–Ω–∞–∫–∏*  
* *–î–∞–Ω–Ω—ã–µ —Ä–∞–∑–±–∏—Ç—ã –Ω–∞ —Ç—Ä–∏ –≤—ã–±–æ—Ä–∫–∏: –æ–±—É—á–∞—é—â—É—é, —Ç–µ—Å—Ç–æ–≤—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é.*  
* *–£–¥–∞–ª–µ–Ω—ã –ª–∏—à–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏*  
* *–ü—Ä–∏–∑–Ω–∞–∫–∏, –∏–º–µ—é—â–∏–µ —Ä–∞–∑–Ω—ã–π –º–∞—Å—à—Ç–∞–± - –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω—ã*

## –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–û–±—É—á–∏–º –º–æ–¥–µ–ª–∏ DecisionTreeClassifier, RandomForestClassifier –∏ LogisticRegression –ø–æ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –∏ –ø–æ–¥–±–µ—Ä–µ–º –ª—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ f1 –º–µ—Ä—ã –¥–ª—è –∫–∞–∂–¥–æ–π.

### –ú–æ–¥–µ–ª—å —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞

```python
%%time

# –°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –ø–æ–∏—Å–∫–∞ –¥–µ—Ä–µ–≤–∞ —Å –ª—É—á—à–µ–π f1 –º–µ—Ä–æ–π
best_f1_tree = 0

# –í —Ü–∏–∫–ª–µ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –∏ –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –∏ f1 –º–µ—Ä—É –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –æ—Ç 2 –¥–æ 40
for depth in range(2, 40,1):
    model_tree = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree.fit(features_train,target_train)
    predicted_valid_tree = model_tree.predict(features_valid)
    f1_tree_score = f1_score(target_valid, predicted_valid_tree)

# –ó–∞–ø–∏—à–µ–º –ª—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ f1 –º–µ—Ä—ã –∏ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞
    if best_f1_tree < f1_tree_score:
        best_f1_tree = f1_tree_score
        best_tree_depth = depth
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_tree}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_tree_depth}</font>"))
```

### –ú–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

```python
%%time

# –°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –ø–æ–∏—Å–∫–∞ –¥–µ—Ä–µ–≤–∞ —Å –ª—É—á—à–µ–π f1 –º–µ—Ä–æ–π
best_f1_forest = 0

# –í —Ü–∏–∫–ª–µ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –∏ –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –∏ f1 –º–µ—Ä—É –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –æ—Ç 2 –¥–æ 20 –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ä–µ–≤—å–µ–≤ –æ—Ç 20 –¥–æ 100
for depth in range(14,20,1):
    for est in range(70,85,1):
        model_forest = RandomForestClassifier(random_state=12345, max_depth=depth, n_estimators=est) 
        model_forest.fit(features_train,target_train)
        predicted_valid_forest = model_forest.predict(features_valid)
        f1_forest_score = f1_score(target_valid, predicted_valid_forest)
     

    # –ó–∞–ø–∏—à–µ–º –ª—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ f1 –º–µ—Ä—ã, –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–µ—Ä–µ–≤—å–µ–≤
        if best_f1_forest < f1_forest_score:
            best_f1_forest = f1_forest_score
            best_forest_depth = depth
            best_forest_est = est
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_forest}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_forest_depth}</font>"))
display(HTML(f"<font color='orange'>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ –Ω–∞–∏–ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å: {best_forest_est}</font>"))
```

### –ú–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

```python
%%time

#–û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
model_log_reg = LogisticRegression(random_state=12345,solver='liblinear')
model_log_reg.fit(features_train,target_train) 
predicted_valid_log = model_log_reg.predict(features_valid)

#–ü–æ—Å—á–∏—Ç–∞–µ–º f1 –º–µ—Ä—É –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
f1_log_score = f1_score(target_valid, predicted_valid_log)


display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{f1_log_score}</font>"))

```

### –†–∞—Å—á—ë—Ç –º–µ—Ç—Ä–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏

*–†–∞—Å—á–∏—Ç–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π: precision_score, recall_score, accuracy_score, roc_auc_score*

```python
# accuracy –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
accuracy = [
    accuracy_score(target_valid,predicted_valid_tree), 
    accuracy_score(target_valid,predicted_valid_forest),
    accuracy_score(target_valid,predicted_valid_log)
]

# recall –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏            
recall = [
    recall_score(target_valid,predicted_valid_tree),
    recall_score(target_valid,predicted_valid_forest),
    recall_score(target_valid,predicted_valid_log)
]

# precision –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
precision = [
    precision_score(target_valid,predicted_valid_tree),
    precision_score(target_valid,predicted_valid_forest),
    precision_score(target_valid,predicted_valid_log)
]

# –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞
probabilities_valid_tree = model_tree.predict_proba(features_valid)  # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_tree = probabilities_valid_tree[:, 1]        # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
probabilities_valid_forest = model_forest.predict_proba(features_valid) # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_forest = probabilities_valid_forest[:, 1]       # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
probabilities_valid_log = model_log_reg.predict_proba(features_valid)   # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_log = probabilities_valid_log[:, 1]             # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# auc_roc –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
auc_roc = [
    roc_auc_score(target_valid,probabilities_one_valid_tree),
    roc_auc_score(target_valid,probabilities_one_valid_forest),
    roc_auc_score(target_valid,probabilities_one_valid_log)
]

f1 = [
    best_f1_tree,
    best_f1_forest,
    f1_log_score    
]

index_name = ['–û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:','–ü–æ–ª–Ω–æ—Ç–∞:','–¢–æ—á–Ω–æ—Å—Ç—å:', 'AUC-ROC:', 'F1-–º–µ—Ä–∞:']

columns_name = ['–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ', '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å', '–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è']

metrics = pd.DataFrame([accuracy, recall, precision, auc_roc, f1], columns = columns_name, index = index_name)

metrics
```

```python
# –ø—Ä–æ–≤–µ—Ä–∏–º, –∫–∞–∫ —á–∞—Å—Ç–æ –≤ —Ü–µ–ª–µ–≤–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –∫–ª–∞—Å—Å ¬´1¬ª –∏–ª–∏ ¬´0¬ª
# –∏ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏

class_frequency = churn_data_ohe['exited'].value_counts(normalize=True)
display(class_frequency)
class_frequency.plot(kind='bar')
plt.show()
```

*–ü—Ä–æ–≤–µ–¥—ë–º –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫—É –≤—ã–±–æ—Ä–∫–∏ —á–µ—Ä–µ–∑ –µ—ë —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º –∏ —Å—Ä–∞–≤–Ω–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ —Ä–∞–Ω–µ–µ.*

### –£–≤–µ–ª–∏—á–µ–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ (upsample)

```python
def upsample(features, target, repeat):
    
    '''
    
    –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–∏–∑–Ω–∞–∫–∏ target, futures, –∞ —Ç–∞–∫ –∂–µ –º–Ω–æ–∂–∏—Ç–µ–ª—å repeat.
    –§—É–Ω–∫—Ü–∏—è —Ä–∞–∑–¥–µ–ª—è–µ—Ç –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ, —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—ã–π –º–Ω–æ–∂–∏—Ç–µ–ª—å, 
    —Å–æ–±–∏—Ä–∞–µ—Ç –æ–±—Ä–∞—Ç–Ω–æ –¥–∞–Ω–Ω—ã–µ –≤ –≤—ã–±–æ—Ä–∫—É –∏ –ø–µ—Ä–µ–º–µ—à–∏–≤–∞–µ—Ç –∏—Ö. 
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É —Å —É–≤–µ–ª–∏—á–µ–Ω–Ω–æ–π –¥–æ–ª–µ–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.
    
    '''
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled


features_upsampled, target_upsampled = upsample(features_train, target_train, 4)  #–£–∫–∞–∑—ã–≤–∞–µ–º –º–Ω–æ–∂–∏—Ç–µ–ª—å 4, —Ç.–∫. –∏—Å—Ö–æ–¥–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ 80%/20% = 4 

display(HTML(f"<font color='red'>–†–∞–∑–º–µ—Ä –Ω–æ–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ features_upsampled: {features_upsampled.shape}</font>"))
display(HTML(f"<font color='blue'>–†–∞–∑–º–µ—Ä –Ω–æ–≤–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ target_upsampled: {target_upsampled.shape}</font>"))

display(HTML(f"<font color='red'>–†–∞–∑–º–µ—Ä —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ features_train: {features_train.shape}</font>"))
display(HTML(f"<font color='blue'>–†–∞–∑–º–µ—Ä —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ target_train: {target_train.shape}</font>"))
```

#### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –Ω–∞ –Ω–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ

##### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
%%time

# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –î–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –Ω–∞ –Ω–æ–≤–æ–π, —É–≤–µ–ª–∏—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
# –ø–æ—Å—á–∏—Ç–∞–µ–º f1 –º–µ—Ä—É –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ 

best_f1_tree_up = 0


for depth in range(2, 40,1):
    model_tree_up = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree_up.fit(features_upsampled,target_upsampled)
    predicted_valid_tree_up = model_tree_up.predict(features_valid)
    f1_tree_score_up = f1_score(target_valid, predicted_valid_tree_up)
    
    if best_f1_tree_up < f1_tree_score_up:
        best_f1_tree_up = f1_tree_score_up
        best_tree_depth_up = depth
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_tree_up}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_tree_depth_up}</font>")) 
```

##### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
%%time

# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –Ω–∞ –Ω–æ–≤–æ–π, —É–≤–µ–ª–∏—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
# –ø–æ—Å—á–∏—Ç–∞–µ–º f1 –º–µ—Ä—É –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ 

best_f1_forest_up = 0

for depth in range(16,20,1):
    for est in range(268,272,1):
        model_forest_up = RandomForestClassifier(random_state=12345, max_depth=depth, n_estimators=est) 
        model_forest_up.fit(features_upsampled,target_upsampled)
        predicted_valid_forest_up = model_forest_up.predict(features_valid)
        f1_forest_score_up = f1_score(target_valid, predicted_valid_forest_up)
     
        
        if best_f1_forest_up < f1_forest_score_up:
            best_f1_forest_up = f1_forest_score_up
            best_forest_depth_up = depth
            best_forest_est_up = est
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_forest_up}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_forest_depth_up}</font>"))
display(HTML(f"<font color='orange'>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ –Ω–∞–∏–ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å: {best_forest_est_up}</font>"))
```

##### –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

```python
%%time

# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —É–≤–µ–ª–∏—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
# –ø–æ–ª—É—á–∏–º –º–µ—Ç—Ä–∏–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ

model_log_reg_up = LogisticRegression(random_state=12345,solver='liblinear')
model_log_reg_up.fit(features_upsampled,target_upsampled) 
predicted_valid_log_up = model_log_reg_up.predict(features_valid)
f1_log_score_up = f1_score(target_valid, predicted_valid_log_up)


display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{f1_log_score_up}</font>"))
```

#### –†–∞—Å—á—ë—Ç –º–µ—Ç—Ä–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏

```python
# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –ø–æ —Ä–∞–Ω–µ–µ –≤—ã—è—Å–Ω–µ–Ω–Ω—ã–º, –Ω–∞–∏–ª—É—á—à–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º
model_tree_best = DecisionTreeClassifier(random_state=12345, max_depth=best_tree_depth_up)
model_tree_best.fit(features_upsampled,target_upsampled)
prediction_tree_best = model_tree_best.predict(features_valid)

# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –ø–æ —Ä–∞–Ω–µ–µ –≤—ã—è—Å–Ω–µ–Ω–Ω—ã–º, –Ω–∞–∏–ª—É—á—à–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º
model_forest_best = RandomForestClassifier(random_state=12345, max_depth=best_forest_depth_up,n_estimators=best_forest_est_up)
model_forest_best.fit(features_upsampled,target_upsampled)
prediction_forest_best = model_forest_best.predict(features_valid)

#–í –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–µ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ—ç—Ç–æ–º—É –æ–±—É—á–∞—Ç—å –∑–∞–Ω–æ–≤–æ –º–æ–¥–µ–ª—å –Ω–µ –Ω—É–∂–Ω–æ

# accuracy –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
accuracy_up = [
    accuracy_score(target_valid,prediction_tree_best), 
    accuracy_score(target_valid,prediction_forest_best),
    accuracy_score(target_valid,predicted_valid_log_up)
]

# recall –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏            
recall_up = [
    recall_score(target_valid,prediction_tree_best),
    recall_score(target_valid,prediction_forest_best),
    recall_score(target_valid,predicted_valid_log_up)
]

# precision –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
precision_up = [
    precision_score(target_valid,prediction_tree_best),
    precision_score(target_valid,prediction_forest_best),
    precision_score(target_valid,predicted_valid_log_up)
]

# –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞
probabilities_valid_tree_up = model_tree_best.predict_proba(features_valid)  # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_tree_up = probabilities_valid_tree_up[:, 1]          # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
probabilities_valid_forest_up = model_forest_best.predict_proba(features_valid) # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_forest_up = probabilities_valid_forest_up[:, 1]         # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
probabilities_valid_log_up = model_log_reg.predict_proba(features_valid)      # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_log_up = probabilities_valid_log_up[:, 1]             # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# auc_roc –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
auc_roc_up = [
    roc_auc_score(target_valid,probabilities_one_valid_tree_up),
    roc_auc_score(target_valid,probabilities_one_valid_forest_up),
    roc_auc_score(target_valid,probabilities_one_valid_log_up)
]

f1_up = [
    best_f1_tree_up,
    best_f1_forest_up,
    f1_log_score_up    
]

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –†–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_tree, tpr_tree, thresholds_tree = roc_curve(target_valid, probabilities_one_valid_tree_up)

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_forest, tpr_forest, thresholds_forest = roc_curve(target_valid, probabilities_one_valid_forest_up)

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–µ–≥—Ä–µ—Å—Å–∏–∏  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_log, tpr_log, thresholds_log = roc_curve(target_valid, probabilities_one_valid_log_up)

index_name = ['–û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:','–ü–æ–ª–Ω–æ—Ç–∞:','–¢–æ—á–Ω–æ—Å—Ç—å:', 'AUC-ROC:', 'F1-–º–µ—Ä–∞:']

columns_name = ['–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ', '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å', '–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è']

metrics_up = pd.DataFrame([accuracy_up, recall_up, precision_up, auc_roc_up, f1_up], columns = columns_name, index = index_name)

metrics_up
```

*–í –≤–∏–¥—É –≤—ã—à–µ—Å–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—Ä–æ–ø—É—Å—Ç–∏–º –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–∞ —É–º–µ–Ω—å—à–µ–Ω–∏—è –≤—ã–±–æ—Ä–∫–∏. –í–º–µ—Å—Ç–æ –Ω–µ–≥–æ –ø—Ä–∏–º–µ–Ω–∏–º –º–µ—Ç–æ–¥ –≤–∑–≤–µ—à–∏–≤–∞–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤.*

#### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤—ã–±–æ—Ä–∫–µ –ø–æ—Å–ª–µ –≤–∑–≤–µ—à–∏–≤–∞–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤

##### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
%%time

# –°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –ø–æ–∏—Å–∫–∞ –¥–µ—Ä–µ–≤–∞ —Å –ª—É—á—à–µ–π f1 –º–µ—Ä–æ–π
best_f1_tree_bal = 0

# –í —Ü–∏–∫–ª–µ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –∏ –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –∏ f1 –º–µ—Ä—É –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –æ—Ç 2 –¥–æ 40
for depth in range(2, 40,1):
    model_tree_bal = DecisionTreeClassifier(random_state=12345, class_weight='balanced', max_depth=depth)
    model_tree_bal.fit(features_train,target_train)
    predicted_valid_tree_bal = model_tree_bal.predict(features_valid)
    f1_tree_score_bal = f1_score(target_valid, predicted_valid_tree_bal)

# –ó–∞–ø–∏—à–µ–º –ª—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ f1 –º–µ—Ä—ã –∏ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞
    if best_f1_tree_bal < f1_tree_score_bal:
        best_f1_tree_bal = f1_tree_score_bal
        best_tree_depth_bal = depth
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_tree_bal:.2f}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_tree_depth_bal}</font>"))
```

##### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
%%time

# –°–æ–∑–¥–∞–¥–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –¥–ª—è –ø–æ–∏—Å–∫–∞ –¥–µ—Ä–µ–≤–∞ —Å –ª—É—á—à–µ–π f1 –º–µ—Ä–æ–π
best_f1_forest_bal = 0

# –í —Ü–∏–∫–ª–µ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –∏ –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –∏ f1 –º–µ—Ä—É –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –æ—Ç 2 –¥–æ 20 –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–µ—Ä–µ–≤—å–µ–≤ –æ—Ç 20 –¥–æ 100
for depth in range(8,12,1):
    for est in range(255,260,1):
        model_forest_bal = RandomForestClassifier(random_state=12345, max_depth=depth, n_estimators=est, 
                                                     class_weight='balanced') 
        model_forest_bal.fit(features_train,target_train)
        predicted_valid_forest_bal = model_forest_bal.predict(features_valid)
        f1_forest_score_bal = f1_score(target_valid, predicted_valid_forest_bal)
     
       # –ó–∞–ø–∏—à–µ–º –ª—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ f1 –º–µ—Ä—ã, –≥–ª—É–±–∏–Ω—ã –¥–µ—Ä–µ–≤–∞ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–µ—Ä–µ–≤—å–µ–≤ 
        if best_f1_forest_bal < f1_forest_score_bal:
            best_f1_forest_bal = f1_forest_score_bal
            best_forest_depth_bal = depth
            best_forest_est_bal = est
display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{best_f1_forest_bal}</font>"))  
display(HTML(f"<font color='blue'>–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞: {best_forest_depth_bal}</font>"))
display(HTML(f"<font color='orange'>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ –Ω–∞–∏–ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å: {best_forest_est_bal}</font>"))
```

##### –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

```python
%%time

# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —É–≤–µ–ª–∏—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
# –ø–æ–ª—É—á–∏–º –º–µ—Ç—Ä–∏–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ

model_log_reg_bal = LogisticRegression(random_state=12345,solver='liblinear', class_weight='balanced')
model_log_reg_bal.fit(features_train,target_train) 
predicted_valid_log_bal = model_log_reg_bal.predict(features_valid)
f1_log_score_bal = f1_score(target_valid, predicted_valid_log_bal)


display(HTML(f"<font color='red'>C—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–Ω–æ—Ç—ã –∏ —Ç–æ—á–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:\
{f1_log_score_bal}</font>"))
```

#### –†–∞—Å—á—ë—Ç –º–µ—Ç—Ä–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏

```python
# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –ø–æ —Ä–∞–Ω–µ–µ –≤—ã—è—Å–Ω–µ–Ω–Ω—ã–º, –Ω–∞–∏–ª—É—á—à–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º
model_tree_best_bal = DecisionTreeClassifier(random_state=12345, class_weight='balanced', max_depth=best_tree_depth_bal)
model_tree_best_bal.fit(features_train,target_train)
prediction_tree_best_bal = model_tree_best_bal.predict(features_valid)

# –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –ø–æ —Ä–∞–Ω–µ–µ –≤—ã—è—Å–Ω–µ–Ω–Ω—ã–º, –Ω–∞–∏–ª—É—á—à–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º
model_forest_best_bal = RandomForestClassifier(random_state=12345, max_depth=best_forest_depth_bal,n_estimators=best_forest_est_bal,
                                              class_weight='balanced')
model_forest_best_bal.fit(features_train,target_train)
prediction_forest_best_bal = model_forest_best_bal.predict(features_valid)

#–í –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–µ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ—ç—Ç–æ–º—É –æ–±—É—á–∞—Ç—å –∑–∞–Ω–æ–≤–æ –º–æ–¥–µ–ª—å –Ω–µ –Ω—É–∂–Ω–æ

# accuracy –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
accuracy_bal = [
    accuracy_score(target_valid,prediction_tree_best_bal), 
    accuracy_score(target_valid,prediction_forest_best_bal),
    accuracy_score(target_valid,predicted_valid_log_bal)
]

# recall –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏            
recall_bal = [
    recall_score(target_valid,prediction_tree_best_bal),
    recall_score(target_valid,prediction_forest_best_bal),
    recall_score(target_valid,predicted_valid_log_bal)
]

# precision –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
precision_bal = [
    precision_score(target_valid,prediction_tree_best_bal),
    precision_score(target_valid,prediction_forest_best_bal),
    precision_score(target_valid,predicted_valid_log_bal)
]

# –¥–ª—è —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞
probabilities_valid_tree_bal = model_tree_best_bal.predict_proba(features_valid)  # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_tree_bal = probabilities_valid_tree_bal[:, 1]          # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
probabilities_valid_forest_bal = model_forest_best_bal.predict_proba(features_valid) # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_forest_bal = probabilities_valid_forest_bal[:, 1]         # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
probabilities_valid_log_bal = model_log_reg_bal.predict_proba(features_valid)      # –û—Ü–µ–Ω–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–æ–≤ –º–µ—Ç–æ–¥–æ–º predict_proba
probabilities_one_valid_log_bal = probabilities_valid_log_bal[:, 1]             # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞

# auc_roc –¥–ª—è –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π, —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
auc_roc_bal = [
    roc_auc_score(target_valid,probabilities_one_valid_tree_bal),
    roc_auc_score(target_valid,probabilities_one_valid_forest_bal),
    roc_auc_score(target_valid,probabilities_one_valid_log_bal)
]

f1_bal = [
    best_f1_tree_bal,
    best_f1_forest_bal,
    f1_log_score_bal    
]

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –†–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_tree_bal, tpr_tree_bal, thresholds_tree_bal = roc_curve(target_valid, probabilities_one_valid_tree_bal)

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –°–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_forest_bal, tpr_forest_bal, thresholds_forest_bal = roc_curve(target_valid, probabilities_one_valid_forest_bal)

#–ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ FPR –∏ TPR –Ω–∞ –º–æ–¥–µ–ª–∏ –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–µ–≥—Ä–µ—Å—Å–∏–∏  –¥–ª—è –≥—Ä–∞–∞—Ñ–∏–∫–∞ AUC-ROC
fpr_log_bal, tpr_log_bal, thresholds_log_bal = roc_curve(target_valid, probabilities_one_valid_log_bal)

index_name = ['–û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:','–ü–æ–ª–Ω–æ—Ç–∞:','–¢–æ—á–Ω–æ—Å—Ç—å:', 'AUC-ROC:', 'F1-–º–µ—Ä–∞:']

columns_name = ['–†–µ—à–∞—é—â–µ–µ –¥–µ—Ä–µ–≤–æ', '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å', '–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è']

metrics_bal = pd.DataFrame([accuracy_bal, recall_bal, precision_bal, auc_roc_bal, f1_bal], columns = columns_name, index = index_name)

metrics_bal
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Å–ª–æ–≤–∞—Ä—å —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ–º –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å
forest_params = {'max_depth': range(16,24),'n_estimators': range(265,272)}

# —Å–æ–∑–¥–∞–¥–∏–º —Å—á–µ—Ç—á–∏–∫ f1 –º–µ—Ä—ã
f1 = make_scorer(f1_score , average='binary')

# —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é –º–æ–¥–µ–ª—å
grid_model = RandomForestClassifier(random_state=12345)

# –ü—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é GridSearch
forest_grid = GridSearchCV(grid_model,forest_params,cv=5, n_jobs=-1,scoring=f1,verbose=True)
```
