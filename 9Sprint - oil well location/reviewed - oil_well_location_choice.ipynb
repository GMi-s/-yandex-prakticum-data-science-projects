## Предобработка данных

### Удаление дубликатов

*Поскольку информация о физической природе признаков в задании не раскрывается, то у нас нет возможности понять, какие из дублированных данных являются более релевантными для представленных скважин. Поэтому просто удалим вторые по счёту дубли в данных по id*

```python
def drop_duplicates_by_id(df):
    '''
    Функция удаляет дубликаты, сохраняя первый найденный экземпляр
    '''
    df.drop_duplicates(subset='id', keep='first', inplace=True)
    df.reset_index(drop=True)
```

```python
drop_duplicates_by_id(geo_0)
drop_duplicates_by_id(geo_1)
drop_duplicates_by_id(geo_2)
```

### Выделение признаков и целевого признака

```python
def features_definition(df):
    '''
    Функция выделяет признаки и целевой признак
    '''
    features = df.drop(['id','product'], axis=1)
    target = df['product']
    
    return features,target
```

```python
features_1,target_1 = features_definition(geo_0)
features_2,target_2 = features_definition(geo_1)
features_3,target_3 = features_definition(geo_2)
```

```python
def split_df(features,target,color,region):
    '''
    Принимает названия выборок с признаками и с целевыми признаками, цвет шрифта и название региона на вывод данных  
    Функция делит данные на обучающую и валидационную выборки.
    Возвращает обучающую и тестовую выборки, разделённые по признакам и целевому признаку
    '''
    
    features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, train_size=0.75,random_state=12345)
    
    display(HTML(f"<font color={color}>Размер обучающей и валидационной выборок с целевыми признаками {region}: \
    {target_train.shape},{target_valid.shape}</font>"))
    display(HTML(f"<font color={color}>Размер с обучающей и валидационной выборок с признаками {region}: \
    {features_train.shape},{features_valid.shape}</font>"))

    return features_train,features_valid,target_train,target_valid
```

```python
def model_linear(features,target,features_train, features_valid, target_train, target_valid, color, region):
    '''
    Функция обучает модель LinearRegression 
    и вычисляет средний запас предсказанного сырья и RMSE
    '''
    model = LinearRegression()
    model.fit(features_train,target_train)
    predictions_valid = model.predict(features_valid)
    result_rmse = mean_squared_error(target_valid,predictions_valid)**0.5
    display(HTML(f"<font color = {color}>Средний запас предсказанного сырья {region}:\
    {predictions_valid.mean():.2f} тысяч баррелей</font>"))
    display(HTML(f"<font color = {color}>Коэффициент RMSE модели {region}:\
    {result_rmse:.2f}</font>"))
    
    return predictions_valid,result_rmse,model
```

### ROI (Return On Investments) - "Возврат на инвестиций"

*ROI скважины = (Доход от скважины - Инвестиции)/Инвестиции. Если ROI равен 0, то вложения окупились, но прибыли нет, а значит это и есть точка безубыточности. Таким образом для точки безубыточности:*<br>
- *Доход от скважины = Инвестиции (`BUDGET_ON_BEST_UNIT`).*  
- *Доход от скважины = Объем сырья х Прибыль от продажи единицы сырья(`REVENUE_PER_BAR`).* <br> 

 *Следовательно:*
- *Объем сырья = Инвестиции / Прибыль от продажи единицы сырья.*

```python
VOLUME_OF_RAW = BUDGET_ON_BEST_UNIT/REVENUE_PER_BAR
display(HTML(f"<font color = 'purple'>Минимальный объём сырья для безубыточной разработки новой скважины: \
    {VOLUME_OF_RAW:.2f} тысяч баррелей</font>"))
```

### Функция расчета прибыли и рисков методом Bootstrap

```python
def profit_and_risks(target,probabilities,color,region):
    '''
    Функция расчета прибыли и рисков методом Bootstrap.
    1) Создаём 1000 рандомных выборок по 500 скважин(исследуемых при разведки региона)
    2) Выбраем для каждой из выборок 200 лучших скважин, применив функцию profit
    3) Находим среднюю прибыль, 95%-й доверительный интервал и риск убытков
    '''
    state = np.random.RandomState(12345)
    values = []
    
    for i in range(1000):
        
        target_subsample = target.sample(n = WELLS_REGION, replace=True, random_state=state) 
        probs_subsample = probabilities[target_subsample.index]
    
        
        prof = profit(target_subsample, probs_subsample)            #С внешней функцией profit
        values.append(prof)                                         #С внешней функцией profit
        
        
#         data_test = pd.concat([probs_subsample, target_subsample],axis=1)                            #Без внешней функции profit
#         data_test.columns = ['predictions','target']                                                 #Без внешней функции profit
#         reg_wells = data_test.sort_values(by = 'predictions',ascending=False).head(BEST_WELLS)       #Без внешней функции profit     
#         quantity_over = (target[reg_wells.index][:BEST_WELLS] - VOLUME_OF_RAW).sum()*REVENUE_PER_BAR #Без внешней функции profit        
#         values.append(quantity_over)                                                                 #Без внешней функции profit

    values = pd.Series(values)
    lower = values.quantile(0.025)/10e6
    upper = values.quantile(0.975)/10e6
    mean = values.mean()/10e6
    
    risks = (values < 0).mean()
    
    display(HTML(f"<font color = {color}>Средняя выручка в {region}:\
    {mean:.2f} млн. руб.</font>"))
    display(HTML(f"<font color = {color}>95% доверительный интервал в {region} от {lower:.2f} до {upper:.2f} \
    млн. руб.</font>"))
    display(HTML(f"<font color = {color}>Вероятность убытков в {region}:\
    {risks:.1%}</font>"))
    
    return mean,lower,upper,risks
```
