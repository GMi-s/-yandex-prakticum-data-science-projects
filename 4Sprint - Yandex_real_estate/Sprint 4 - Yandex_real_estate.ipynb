## 2. Предобработка данных

###  Обработка пропусков. Фильтрация значений.

Согласно сделанным выше выводам поэтапно выполним дополнительные исследования и устраним выявленные замечания.

```python
real_estate_data.columns = ['total_images', 'last_price', 'total_area', 'first_day_exposition',
       'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor',
       'is_apartment', 'studio', 'open_plan', 'kitchen_area', 'balcony',
       'locality_name', 'airports_nearest', 'city_centers_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'days_exposition']
```

Проведём дополнительный анализ категории last_price с целью исключить неправдоподобно малые и неправдоподобно большие значения, а так же изучим причину существенного различия между медианным и средним значениями.

```python
import matplotlib.pyplot as plt

print (real_estate_data['last_price'].describe())
plt.ylim(0, 3e+07)
real_estate_data.boxplot (column = 'last_price')
#real_estate_data['last_price'].hist(bins=100)
```

Анализ показывает, что половина значений цены лежит в пределах 3,4 .. 6,8 млн.р. Однако, Распределение имеет большой "хвост" в сторону больших значений.<br>
Проанализируем данные в зонах минимальных и максимальных значений.

```python
print('Доля недвижимости стоимостью дешевле 1 млн.р.:{:.2%}'.format(len(real_estate_data[real_estate_data['last_price'] < 1e+06]) / len(real_estate_data['last_price'])))
print('Доля недвижимости стоимостью более 100 млн.р.:{:.2%}'.format(len(real_estate_data[real_estate_data['last_price'] > 1e+08]) / len(real_estate_data['last_price'])))
print(real_estate_data[real_estate_data['last_price'] < 1e+06].sort_values(by = 'last_price').head())
```

Доля недвижимости стоимостью дешевле 1 млн.р. составляет менее 1%. Однако с учётом цен в удалённых населённых пунктах только одно значение в Санкт-Петербурге стоимостью 12190.0 р. выглядит неправдоподобно. Скорее всего данная цена является арендной и поэтому данное значение будет отфильтровано. <br>
Доля недвижимости стоимостью более 100 млн.р. составляет менее 0,16%, что несмотря на высокую стоимость, так же можно считать правдоподобным в случае, если такие объекты относятся к категории элитной недвижимости. <br>
Далее удобства дальнейшей работы изменим порядок единиц измерения из руб. в млн. руб.

```python
real_estate_data = real_estate_data[real_estate_data['last_price'] != 12190.0]
real_estate_data['last_price'] = real_estate_data['last_price'] / 1e+06
```

## 3. Расчёты. Добавление данных в таблицу.

Добавим в общую таблицу дополнительную категорию распределения предложений о продаже по этажам: "первый", "последний", "другой"

```python
def floor_category (row):

    '''
    
    Возвращает категорию расположения жилья по этажам, используя категории:
    - 'первый', если значение в категории 'floor' == 1
    - 'последний', если значение в категории 'floor' == 'floors_total'
    - 'другой', все остальные значения
    
    '''


    floor = row['floor']
    top_floor = row['floors_total']
    
    if floor == 1:
        return 'первый'
    
    if floor == top_floor:
        return 'последний'
    
    return 'другой'


real_estate_data ['floor_category'] = real_estate_data.apply(floor_category, axis = 1)

```

При этом размещением объявлений занимаются не только частные лица, но и риэлтерские агентства, которые быстро наполнили систему своими объявлениями о продаже дорогих, трудно продаваемых квартир, что повлияло на расчёт медианной цены. #### *Выберите 10 населённых пунктов с наибольшим числом объявлений. Посчитайте среднюю цену квадратного метра в этих населённых пунктах. Выделите среди них населённые пункты с самой высокой и низкой стоимостью жилья. Эти данные можно найти по имени в столбце locality_name.*

```python
#Создадим сводную таблицу по названиями населённых пунктов и количеству объявлений о продаже квартир в них, и отсортируем её в порядке убывания кол-ва объявлений
top_ten_offers_qty_locations = real_estate_data.pivot_table(index = 'locality_name', \
    values = 'last_price', aggfunc = 'count').sort_values(by = 'last_price', ascending = False)

#Создадим пустой список в который будем добавлять значения средних стоимостей за кв. м. в каждом населённом пункте
price_per_meter_list = []

#Для каждого населённого пункта из сводной таблицы найдём в основной таблице все значения стоимостей за кв. м. и рассчитаем медиану
for row in top_ten_offers_qty_locations.index:
    condition_list = real_estate_data[real_estate_data['locality_name'] == row]
    price_per_meter_by_location = condition_list['price_per_meter'].median()
    price_per_meter_list.append (price_per_meter_by_location)

#Добавим данные о средней стоимости за кв. м. для каждого населённого пункта в сводную таблицу
top_ten_offers_qty_locations['price_per_meter'] = price_per_meter_list

#Переименуем столбцы для удобства
top_ten_offers_qty_locations.columns = ['количество объявлений', 'цена за кв. метр']

#Переиндексируем таблицу для фильтрации первых 10 значений в соответствии с задачей и отфильтруем все, кроме первых 10 значений.
top_ten_offers_qty_locations = top_ten_offers_qty_locations.reset_index()
top_ten_offers_qty_locations = top_ten_offers_qty_locations.loc[top_ten_offers_qty_locations.index <= 10,:]

top_ten_offers_qty_locations_max_price = top_ten_offers_qty_locations[top_ten_offers_qty_locations['цена за кв. метр'] == \
                                                                     top_ten_offers_qty_locations['цена за кв. метр'].max()]
top_ten_offers_qty_locations_min_price = top_ten_offers_qty_locations[top_ten_offers_qty_locations['цена за кв. метр'] == \
                                                                     top_ten_offers_qty_locations['цена за кв. метр'].min()]
print ('Населённый пункт с максимальной средней ценой за кв.м. и 10 населённых пунктов с наибольшим количеством объявлений:\n',\
       top_ten_offers_qty_locations_max_price,'\n','\n', \
       'Населённый пункт с минимальной средней ценой за кв.м. и 10 населённых пунктов с наибольшим количеством объявлений:\n',\
       top_ten_offers_qty_locations_min_price)


top_ten_offers_qty_locations
```

Из всех объявлений наибольшее количество объявлений относится к Санкт-Петербургу. В нём же и самая большая средняя цена за кв. метр. На 10-й строчке рейтинга по количеству объявлений расположился Петергоф, но наименьшая цена за квадратный метр в Выборге, который занимает 9-ю строчку рейтинга.<br>
