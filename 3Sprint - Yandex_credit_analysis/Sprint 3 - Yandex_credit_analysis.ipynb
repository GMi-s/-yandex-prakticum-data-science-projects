Однако данный столбец содержит значение 0, которое противоречит здравому смыслу и, скорее всего, было подставлено программой автоматически в результате пропуска информации в данном поле при ручном вводе. Разобъём заёмщков на группы по возрастам и заменим пропуски в столбце 'total_income'на медианные значения для каждой группы.

```python
print(len(credit_data[credit_data['total_income'].isna()]))

index = credit_data.index
age_groups = [[25, 30], [30, 35], [35, 40], [40, 45], [45, 50], [50, 55], [55, 60], [60, 65]]
for group in age_groups:
    credit_data.loc[(credit_data['dob_years'] > group[0]) & (credit_data['dob_years'] <= group[1]) \
                    & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] > group[0]) \
                  & (credit_data['dob_years'] <= group[1])]['total_income'].median()
    
    condition = credit_data[(credit_data['dob_years'] > group[0]) & (credit_data['dob_years'] <= group[1])].index
    credit_data.loc[condition,'mean_age'] = (group[0] + group[1]) / 2

    
credit_data.loc[(credit_data['dob_years'] <= 25) & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] <= 25)]['total_income'].median()
credit_data.loc[credit_data['dob_years'] <= 25,'mean_age'] = credit_data[credit_data['dob_years'] <= 25]['dob_years'].mean()
 
credit_data.loc[(credit_data['dob_years'] > 65) & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] > 65)]['total_income'].median()
credit_data.loc[credit_data['dob_years'] > 65,'mean_age'] = credit_data[credit_data['dob_years'] > 65]['dob_years'].mean()

credit_data['mean_age'] = credit_data['mean_age'].astype('int')

#print (credit_data['mean_age'].sort_values().unique())

print(len(credit_data[credit_data['total_income'].isna()]))
#print (credit_data)
```

Аналогичным образом поступим с пропущенными данными в столбце 'days_employed'. Заменим их медианными значениями, разбив заёмщиков на группы по типу дохода 'income_type'. Но т.к. в этой категории встречаются и отрицательные, и положительные значения, последние из которых, по всей видимости, выражены не в днях, а в часах, то сначала переведём все значения их часов в дни, а затем сделаем все значения положительными.

### Лемматизация

Раздел данных 'purpose' содержит множество неявных дубликатов в следствие ручного ввода. Ознакомимся со списком уникальных значений и применим лемматизацию к данной категории.

```python
print (credit_data['purpose'].unique())
```

```python
from pymystem3 import Mystem
m = Mystem()
from collections import Counter
 
lem_list = [] 
for word in credit_data['purpose']:
    lemmas = ' '.join(m.lemmatize(word))
    lem_list.append(lemmas)
#    print(lemmas)
#    print(Counter(lemmas))

credit_data['lemmas'] = lem_list
#print (lem_list)
#print (credit_data)
print (pd.Series(lem_list).unique())
```
