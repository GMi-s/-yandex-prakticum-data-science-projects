```python
print(len(credit_data[credit_data['total_income'].isna()]))

index = credit_data.index
age_groups = [[25, 30], [30, 35], [35, 40], [40, 45], [45, 50], [50, 55], [55, 60], [60, 65]]
for group in age_groups:
    credit_data.loc[(credit_data['dob_years'] > group[0]) & (credit_data['dob_years'] <= group[1]) \
                    & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] > group[0]) \
                  & (credit_data['dob_years'] <= group[1])]['total_income'].median()
    
    condition = credit_data[(credit_data['dob_years'] > group[0]) & (credit_data['dob_years'] <= group[1])].index
    credit_data.loc[condition,'mean_age'] = (group[0] + group[1]) / 2

    
credit_data.loc[(credit_data['dob_years'] <= 25) & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] <= 25)]['total_income'].median()
credit_data.loc[credit_data['dob_years'] <= 25,'mean_age'] = credit_data[credit_data['dob_years'] <= 25]['dob_years'].mean()
 
credit_data.loc[(credit_data['dob_years'] > 65) & (credit_data['days_employed'].isna() == True),'total_income'] \
    = credit_data[(credit_data['dob_years'] > 65)]['total_income'].median()
credit_data.loc[credit_data['dob_years'] > 65,'mean_age'] = credit_data[credit_data['dob_years'] > 65]['dob_years'].mean()

credit_data['mean_age'] = credit_data['mean_age'].astype('int')

#print (credit_data['mean_age'].sort_values().unique())

print(len(credit_data[credit_data['total_income'].isna()]))
#print (credit_data)
```

## –®–∞–≥ 3. –û—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã

- –ï—Å—Ç—å –ª–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –Ω–∞–ª–∏—á–∏–µ–º –¥–µ—Ç–µ–π –∏ –≤–æ–∑–≤—Ä–∞—Ç–æ–º –∫—Ä–µ–¥–∏—Ç–∞ –≤ —Å—Ä–æ–∫?

```python
cd_children_pivot = credit_data.pivot_table(index=['children'], columns='debt', values='total_income', aggfunc='count')

for index in cd_children_pivot:
    
    try:
        cd_children_pivot['ratio'] = cd_children_pivot [0] / cd_children_pivot [1]
        
    except:
        cd_children_pivot['ratio'] = NaN

final_children_presence_ratio = cd_children_pivot.loc[1:,0].sum() / cd_children_pivot.loc[1:,1].sum()

print (cd_children_pivot)
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤–æ–∑–≤—Ä–∞—â—ë–Ω–Ω—ã—Ö –≤ —Å—Ä–æ–∫ –∫—Ä–µ–¥–∏—Ç–æ–≤ –∫ –Ω–µ–≤–æ–∑–≤—Ä–∞—â—ë–Ω–Ω—ã–º –≤ —Å—Ä–æ–∫ –¥–ª—è –∑–∞—ë–º—â–∏–∫–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π', cd_children_pivot.loc[0,'ratio'])
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤–æ–∑–≤—Ä–∞—â—ë–Ω–Ω—ã—Ö –≤ —Å—Ä–æ–∫ –∫—Ä–µ–¥–∏—Ç–æ–≤ –∫ –Ω–µ–≤–æ–∑–≤—Ä–∞—â—ë–Ω–Ω—ã–º –≤ —Å—Ä–æ–∫ –¥–ª—è –∑–∞—ë–º—â–∏–∫–æ–≤ —Å –¥–µ—Ç—å–º–∏', final_children_presence_ratio)
```

    <br>
    –ù–æ–≤—ã–π —Ä–∞—Å—á—ë—Ç –Ω–∏–∂–µ:

```python
cd_children_pivot_narrow  = credit_data.groupby(['children','debt']).agg({'total_income':'count'})
print (cd_children_pivot_narrow)

cd_children_pivot_narrow = cd_children_pivot_narrow.reset_index(level = 'children') #—Ä–µ–∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –ø–æ –Ω–∞–ª–∏—á–∏—é –¥–æ–ª–≥–∞ –∏ –¥–µ—Ç–µ–π

#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π –∏ —Å –¥–æ–ª–≥–æ–º / (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π –∏ —Å –¥–æ–ª–≥–æ–º + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π –∏ –±–µ–∑ –¥–æ–ª–≥–∞)
no_children_debt_to_total_no_children_ratio = cd_children_pivot_narrow[(cd_children_pivot_narrow.index == 1) & \
                             (cd_children_pivot_narrow['children'] == 0)]['total_income'].sum() / \
                                cd_children_pivot_narrow[cd_children_pivot_narrow['children'] == 0]\
                                    ['total_income'].sum()

#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –¥–µ—Ç—å–º–∏ —Å –¥–æ–ª–≥–∞–º–∏ / (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –¥–µ—Ç—å–º–∏ –∏ —Å –¥–æ–ª–≥–∞–º–∏ + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –¥–µ—Ç—å–º–∏ –±–µ–∑ –¥–æ–ª–≥–æ–≤)
with_children_debt_to_total_with_children_ratio = \
    cd_children_pivot_narrow[(cd_children_pivot_narrow.index == 1) & \
                             (cd_children_pivot_narrow['children'] > 0)]['total_income'].sum() / \
                                cd_children_pivot_narrow[cd_children_pivot_narrow['children'] > 0]\
                                    ['total_income'].sum() 

print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª-–≤–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π —Å –¥–æ–ª–≥–∞–º–∏ –∫ –æ–±—â–µ–º—É –∫–æ–ª-–≤—É –∫–ª–∏–µ–Ω—Ç–æ–≤ –±–µ–∑ –¥–µ—Ç–µ–π {:.2%}'.format (no_children_debt_to_total_no_children_ratio))
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª-–≤–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ c –¥–µ—Ç—å–º–∏ —Å –¥–æ–ª–≥–∞–º–∏ –∫ –æ–±—â–µ–º—É –∫–æ–ª-–≤—É –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –¥–µ—Ç—å–º–∏ {:.2%}'.format (with_children_debt_to_total_with_children_ratio))

#print(cd_children_pivot_narrow)

```

```python
#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∏–∑–∫–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ / (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∏–∑–∫–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∏–∑–∫–∏–º –¥–æ—Ö–¥–æ–º –±–µ–∑ –¥–æ–ª–≥–æ–≤)
low_income_rank_debt_clients_ratio = \
 cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–Ω–∏–∑–∫–∏–π', 1].sum() / \
 (cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–Ω–∏–∑–∫–∏–π', 0].sum() + \
  cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–Ω–∏–∑–∫–∏–π', 1].sum())

#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ / (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –¥–æ—Ö–¥–æ–º –±–µ–∑ –¥–æ–ª–≥–æ–≤)
medium_income_rank_debt_clients_ratio = \
 cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '—Å—Ä–µ–¥–Ω–∏–π', 1].sum() / \
 (cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '—Å—Ä–µ–¥–Ω–∏–π', 0].sum() + \
  cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '—Å—Ä–µ–¥–Ω–∏–π', 1].sum())

#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ / (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º –¥–æ—Ö–æ–¥–æ–º —Å –¥–æ–ª–≥–∞–º–∏ + –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º –¥–æ—Ö–¥–æ–º –±–µ–∑ –¥–æ–ª–≥–æ–≤)
high_income_rank_debt_clients_ratio = \
 cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–≤—ã—Å–æ–∫–∏–π', 1].sum() / \
 (cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–≤—ã—Å–æ–∫–∏–π', 0].sum() + \
  cd_income_rank_pivot.loc[cd_income_rank_pivot['income_rank'] == '–≤—ã—Å–æ–∫–∏–π', 1].sum())



#print (cd_income_rank_pivot)
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª-–≤–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∏–∑–∫–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ —Å –¥–æ–ª–≥–∞–º–∏ –∫ –æ–±—â–µ–º—É –∫–æ–ª-–≤—É –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∏–∑–∫–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ {:.2%}'.format (low_income_rank_debt_clients_ratio))
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª-–≤–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ —Å –¥–æ–ª–≥–∞–º–∏ –∫ –æ–±—â–µ–º—É –∫–æ–ª-–≤—É –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–æ —Å—Ä–µ–¥–Ω–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ {:.2%}'.format (medium_income_rank_debt_clients_ratio))
print ('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª-–≤–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ —Å –¥–æ–ª–≥–∞–º–∏ –∫ –æ–±—â–µ–º—É –∫–æ–ª-–≤—É –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º —É—Ä–æ–≤–Ω–µ–º –¥–æ—Ö–æ–¥–∞ {:.2%}'.format (high_income_rank_debt_clients_ratio))                           
```
